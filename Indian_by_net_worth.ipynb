```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_Indians_by_net_worth"
page=requests.get(url)
soup = BeautifulSoup(page.content,'lxml')
soup
```

```python
data = soup.find_all(class_="wikitable sortable")[0]
data
```

```python
heads=[h.get_text().strip().replace('\n',' ') for h in data.find_all('th')]
heads
```

```python
tdata = data.find_all('tr')[1:]
tdata
```

```python
wealth_ch=[]
for i in tdata:
    
    wealth=i.find_all('td')[2].find('img')
    if i.find_all('td')[2].find('img'):
        wealth_ch.append(wealth['alt'])
    else:
        wealth_ch.append(i.find_all('td')[2].get_text().strip())
    
wealth_ch
```

```python
res=[[t.get_text().strip() for t in i.find_all('td')] for i in tdata]
res
```

```python
res[0]
```

```python
heads
```

```python
len(res), len(wealth_ch)
```

```python
for i in range(len(res)):
    res[i][2] = wealth_ch[i]
res
```

```python
d={k:[] for k in heads}
for t in res:
    d['Rank'].append(t[0])
    d['Name'].append(t[1])
    d['Wealth Change'].append(t[2])
    d['Net worth (USD)'].append(t[3])
    d['Company'].append(t[4])
    d['Sources of wealth'].append(t[5])
d
```

```python
df = pd.DataFrame(d)
df
```

```python
df.to_csv('billioniors.csv', index=False)
```
